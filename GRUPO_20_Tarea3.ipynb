```python
from scipy.stats import binom
import matplotlib.pyplot as plt
import numpy as np
```

# La moneda tiene las mismas probabilidad de caer cara o sello

## Probabilidad de obtener exactamente 7 caras en 10 lanzamientos

```python
lanzamientos = 10  # Número de lanzamientos
p_caras = 0.5  # Probabilidad de caer cara
n_caras = 7  # Obtener 7 caras

# Probabilidad de obtener 7 caras
probabilidad_7_caras = binom.pmf(n_caras, lanzamientos, p_caras)

print(f"La probabilidad de obtener {n_caras} caras en {lanzamientos} lanzamientos es de {probabilidad_7_caras:.4f}")
```

### Probabilidad acumulada

```python
lanzamientos = 10  # Número de lanzamientos
p_caras = 0.5  # Probabilidad de caer cara
n_caras = 7  # Obtener 7 caras

# Probabilidad acumulada de obtener 7 o menos caras
probabilidad_acumulada = binom.cdf(n_caras, lanzamientos, p_caras)

print(f"La probabilidad acumulada de obtener {n_caras} o menos caras en {lanzamientos} lanzamientos es de {probabilidad_acumulada:.4f}")
```

### Gráfica de la distribución binomial

```python
lanzamientos = 10  # Lanzamientos
p_caras = 0.5  # Probabilidad de caer cara
valores = np.arange(0, lanzamientos+1)  # Valores posibles: de 0 a 10

# Probabilidades para cada valor
probabilidades = binom.pmf(valores, lanzamientos, p_caras)

# Gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(valores, probabilidades, color="skyblue")

plt.xlabel("Cantidad de veces que sale cara")
plt.ylabel("Probabilidad")
plt.title("Distribución binomial de la probabilidad de obtener entre 0 y 10 caras en 10 lanzamientos")
plt.xticks(valores)

plt.show()
```

# Conjunto de datos aleatorios con 1000 muestras a partir de esta distribución

```python
import math
import statistics
import numpy as np
from scipy import stats
from scipy.stats import norm
import scipy
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ad= norm.rvs(size= 1000)
```

```python
mean_ad=ad.mean()
mean_ad
```

```python
std_ad=ad.std()
std_ad
```

```python
def calculate_median(data):
    n = len(data)
    sorted_data = sorted(data)

    if n % 2:
        median_value = sorted_data[round(0.5 * (n - 1))]
    else:
        index = round(0.5 * n)
        median_value = 0.5 * (sorted_data[index - 1] + sorted_data[index])

    return median_value
median_ = calculate_median(ad)
print("Median:", median_)
```

```python
sns.histplot(ad, bins=50, kde=False, color="#B2EBF2", stat="density")

xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean_ad, std_ad)
plt.plot(x, p, 'k', linewidth=2)

plt.title("Histograma de datos + curva de distribución normal")
plt.xlabel("Valor")
plt.ylabel("Densidad")
```
